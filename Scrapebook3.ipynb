```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.legislation.govt.nz/regulation/public/2022/0168/latest/whole.html"
headers = {"user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}
request = requests.get(url, headers=headers)
soup = BeautifulSoup(request.text, "html.parser")
```

Scrape table 1

```python
# Find the div with id "LMS698035"
div_table = soup.find("div", {"id": "LMS698035"})

table_body = div_table.find("tbody")
table_rows = table_body.find_all("tr")
```

```python
# Create an empty list to hold the table data
table_data = []

# Iterate over each table row
for row in table_rows:
    # Extract data from each table cell
    cells = row.find_all(["td", "th"])  # Modified to include th tags as well
    row_data = [cell.get_text(strip=True) for cell in cells]
    # Append row data to the table data list
    table_data.append(row_data)
```

```python
# Convert the table data to a DataFrame
df = pd.DataFrame(table_data)

# Print the DataFrame
df
```

Do some wrangling

```python
# Drop unnecessary columns
df_clean = df.drop([1, 3], axis=1)
df_clean

# Rename remaining columns
table1_df = df_clean.rename(columns={0: "Determinand", 2: "Maximum Acceptable Value", 4: "Notes"})
table1_df
```
